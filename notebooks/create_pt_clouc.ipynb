```python
from pathlib import Path
import sys; sys.path.append('..')
import scareddataset.data_maniputlation as dm 
from scareddataset import iotools
from scareddataset.calibrator import StereoCalibrator
```

```python
root_path = Path('/home/dimitrisps/new_data/disparity')
out_path = root_path.parent /'final_point_clouds'
out_path.mkdir(parents=True, exist_ok=True)
```

```python
disparities = sorted([str(disparity_path) for disparity_path in root_path.iterdir()])
for p in disparities:
    disparity = iotools.load_subpix_png(p)[0]
    calib = StereoCalibrator().load('/home/dimitrisps/new_data/complete_calib/'+Path(p).stem+'.json')
    # print(calib)
    pts3d = dm.disparity_to_original_scared(disparity, calib)
    iotools.export_ply(out_path/(Path(p).stem+'.ply'), pts3d.reshape(-1,3))
```
